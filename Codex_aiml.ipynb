```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

```

```python
df=pd.read_csv("/kaggle/input/salary-dataset-simple-linear-regression/Salary_dataset.csv")
```

```python
plt.scatter(df['YearsExperience'], df['Salary'])
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot")
plt.show()
```

```python
#step 3 
X = df[['YearsExperience']]
Y = df['Salary']
```

```python
#step 4
model = LinearRegression()
model.fit(X,Y)
slope = model.coef_[0]
intercept = model.intercept_
```

```python
#step 5
def linear_regression(x):
    return slope * x + intercept
```

```python
#step 6 
plt.scatter(df['YearsExperience'], df['Salary'], label='Data Points')
plt.plot(df['YearsExperience'], [linear_regression(xi) for xi in df['YearsExperience']])
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Linear Regression")
plt.show()

```

```python
new_x = 6 
prediction = linear_regression(new_x)
print(f"Prediction for x = {new_x}:{prediction}")

newx = pd.DataFrame({'YearsExperience':[6]})
print(model.predict(newx))
```
